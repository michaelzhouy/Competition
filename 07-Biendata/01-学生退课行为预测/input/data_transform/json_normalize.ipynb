```python
import pandas as pd
import gc
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
```

```python
df = pd.read_json('../Track1/user_video_act_train_4.json', lines=True)
df.iloc[0, :]
```

```python
df.loc[0, 'activity']
```

```python
tmp = df['activity'].apply(pd.json_normalize)
data_list = []
for i in range(len(tmp)):
    d = tmp[i].copy()
    d['item_id'] = df.loc[i, 'id']
    data_list.append(d)
    del d
    gc.collect()

act_4 = pd.concat(data_list, ignore_index=False)
act_4.head()
```

```python
# act_4['item_id'] = ['user_{}'.format(i) for i in range(len(act_4))]
tmp = act_4.groupby(['item_id', 'course_id'])['drop'].apply(list)
tmp.reset_index()
tmp.head()
```

```python
tmp = pd.DataFrame(tmp)
tmp['drop'] = tmp['drop'].map(lambda x: 1 if 1 in x else 0)
tmp['drop'].mean()
```
