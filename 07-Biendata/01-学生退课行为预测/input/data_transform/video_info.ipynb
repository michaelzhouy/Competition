```python
import numpy as np
import pandas as pd
import math
import gc
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
```

```python
df = pd.read_json('../Track1/video_info.json', lines=True)
print(df.shape)
df.head()
```

```python
def video_process(s):
    video_id = s['id']
    name = s['name']
    start = s['start']
    end = s['end']
    text = s['text']
    
    if type(start) != list and math.isnan(float(start)):
        start = [np.nan]
    if type(end) != list and math.isnan(float(end)):
        end = [np.nan]
    if type(text) != list and math.isnan(float(text)):
        text = [np.nan]
    
    frame = pd.DataFrame({
        'video_id': video_id,
        'name': name,
        'start': start,
        'end': end,
        'text': text
    })
    return frame

tmp_df = df.apply(video_process, axis=1)
video_info = pd.concat([i for i in tmp_df], ignore_index=False)
print(video_info.shape)
video_info.head()
```

```python
video_info.to_hdf('../video_info.h5', 'df', index=False)
```
