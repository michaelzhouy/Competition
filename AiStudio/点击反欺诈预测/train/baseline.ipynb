```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_hdf('../input/train.h5')
train.head()
```

```python
test1 = pd.read_hdf('../input/test1.h5')
test1.head()
```

```python
concat_data = pd.concat([train, test1],
                        ignore_index=True)
concat_data.head()
```

```python
concat_data.shape
```

```python
concat_data.info()
```

```python
train_train_X, train_eval_X, train_train_y, train_eval_y = train_test_split(train.drop('label', axis=1), train['label'],
                                                                            random_state=202)
```

```python
id_feat = ['sid', 'android_id']
```

```python
train_cols = [i for i in train.columns if i not in ['timestamp', 'label', 'package', 'media_id', 'fea_hash'] + id_feat]

y = 'label'

lgb_train = lgb.Dataset(train_train_X[train_cols].values, train_train_y) 

lgb_eval= lgb.Dataset(train_eval_X[train_cols].values, train_eval_y, reference=lgb_train)  



params = {

'task': 'train',

'boosting_type': 'gbdt',

'objective': 'binary',

'metric': 'auc',

'num_leaves': 255,

'learning_rate': 0.05,

'feature_fraction': 0.95,

'bagging_fraction': 0.85,

'bagging_freq': 5, 

'min_data_in_leaf':15,

'verbose': 0 

}

print('Start training...')

# train

gbm_val_0 = lgb.train(params,

lgb_train,

num_boost_round=2000,

valid_sets=[lgb_eval, lgb_train],

early_stopping_rounds=50,verbose_eval=10) 

```

```python
test1['label'] = np.where(gbm_val_0.predict(test1[train_cols]) > 0.5, 1, 0)
test1[['sid', 'label']].to_csv('../results/sub_baseline.csv', index=False)
```
