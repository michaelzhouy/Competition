```python
train['srcAddressPort'] = train['srcAddress'].astype(str) + train['srcPort'].astype(str)
train['destAddressPort'] = train['destAddress'].astype(str) + train['destPort'].astype(str)

test['srcAddressPort'] = test['srcAddress'].astype(str) + test['srcPort'].astype(str)
test['destAddressPort'] = test['destAddress'].astype(str) + test['destPort'].astype(str)
```

```python
cat_cols = ['srcAddress', 'destAddress', 'tlsVersion', 'tlsSubject', 'tlsIssuerDn', 'tlsSni', 'srcAddressPort', 'destAddressPort']
```

```python
for i in cat_cols:
    print(i)
    train_unique = train[i].unique()
    test_unique = test[i].unique()
    print('只在test中出现的元素个数：\n', len(set(test_unique) - set(train_unique)))
    print('只在test中出现的元素：\n', set(test_unique) - set(train_unique))
```
