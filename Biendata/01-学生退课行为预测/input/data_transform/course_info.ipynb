```python
import pandas as pd
import gc
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
```

```python
df = pd.read_json('../Track1/course_info.json', lines=True)
df.head()
```

```python
df.shape
```

```python
len(df)
```

```python
tmp = pd.DataFrame({'video_id': df.iloc[0, 0], 'course_id': df.iloc[0, 1]})
tmp
```

```python
course_info = pd.DataFrame()
for i in range(len(df)):
    tmp = pd.DataFrame({'video_id': df.iloc[i, 0], 'course_id': df.iloc[i, 1]})
    course_info = pd.concat([course_info, tmp], ignore_index=False, axis=0)
    del tmp
    gc.collect()
```

```python
course_info.head().append(course_info.tail())
```

```python
course_info.shape
```

```python
course_info.index = range(len(course_info))
course_info.head().append(course_info.tail())
```

```python
# course_info['course_id_maohao'] = course_info['course_id'].map(lambda x: 1 if ':' in x else 0)
course_info['course_id'] = course_info['course_id'].map(lambda x: x.split(':')[1])
gc.collect()
```

```python
course_info.to_csv('../../input/course_info.csv', index=False)
course_info.to_hdf('../../input/course_info.h5', 'df', index=False)
```
