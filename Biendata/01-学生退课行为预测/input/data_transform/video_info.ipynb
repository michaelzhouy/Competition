```python
def course_process(s):
    course_id = s['course_id']
    video_id = s['item']
    
    frame = pd.DataFrame({
        'course_id': course_id,
        'video_id': video_id
    })
    return frame

def video_process(s):
    video_id = s['id']
    name = s['name']
    start = s['start']
    end = s['end']
    text = s['text']
    
    frame = pd.DataFrame({
        'video_id': video_id,
        'name': name,
        'start': start,
        'end': end,
        'text': text
    },index=range(len(start)))
    return frame

tmp_df = df.apply(video_process, axis=1)
video_info = pd.concat([i for i in tmp_df], ignore_index=False)
print(video_info.shape)
video_info.head()
```
