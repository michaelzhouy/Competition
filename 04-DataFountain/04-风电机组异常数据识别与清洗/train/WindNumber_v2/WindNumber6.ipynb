```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.cluster import DBSCAN
from utils import *

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('../../input/dataset.csv')
# sub = pd.read_csv('../../input/submission.csv')
df.head()
```

```python
res = df.loc[df['WindNumber'] == 6, :]
df1 = res.drop_duplicates()

res.shape[0], df1.shape[0]
```

```python
fig = plt.figure()
plt.scatter(df1['WindSpeed'], df1['Power'], s=5, c='r', cmap='viridis')
plt.xlabel('Wind Speed(2)')
plt.ylabel('Power')
plt.xticks(range(0, 26, 2))
plt.grid()
plt.show()
```

```python
df1['Power'].min(), df1['Power'].max()
```

```python
data = df1.copy()

data['low'] = data.apply(lambda x: low_limit(x['WindSpeed'], 3, 16.8, 25, 99, 2000, 0.1), axis=1)
data['high'] = data.apply(lambda x: high_limit(x['WindSpeed'], 99, 2160), axis=1)
# df1['low'].min(), df1['low'].max(), df1['high'].min(), df1['high'].max()

fig = plt.figure()
plt.scatter(data['WindSpeed'], data['Power'], s=3, c='r', cmap='viridis', label='Origin')
plt.scatter(data['WindSpeed'], data['low'], s=3, c='b', cmap='viridis', label='Low')
plt.scatter(data['WindSpeed'], data['high'], s=3, c='k', cmap='viridis', label='High')
plt.xlabel('Wind Speed')
plt.ylabel('Power')
plt.xticks(range(0, 26, 2))
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
data['label'] = data.apply(lambda x: 0 if (x['Power'] > x['low']) and (x['Power'] < x['high']) else 1, axis=1)

fig = plt.figure()
data_0 = data.loc[data['label'] == 0, :]
data_1 = data.loc[data['label'] == 1, :]
plt.scatter(data_0['WindSpeed'], data_0['Power'], s=5, c='r', cmap='viridis', label='0')
plt.scatter(data_1['WindSpeed'], data_1['Power'], s=5, c='b', cmap='viridis', label='1')
plt.xlabel('Wind Speed')
plt.ylabel('Power')
plt.xticks(range(0, 26, 2))
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
db_data = data_0.drop('label', axis=1).sort_values(by='WindSpeed').copy()
print(db_data['WindSpeed'].min(), db_data['WindSpeed'].max())
db_data['WindSpeed_bins'] = pd.cut(db_data['WindSpeed'], bins=list(np.arange(2.5, 18.5, 0.5)), labels=range(1, 32))

data3 = pd.DataFrame()
plt.figure(figsize=(30, 40))
for i in range(1, 32):
#     print('--------{}---------'.format(i))
    bin = db_data.loc[db_data['WindSpeed_bins'] == i, :]
    model = DBSCAN(eps=50, min_samples=5).fit(bin[['WindSpeed', 'Power']])
    # print('labels', i, np.unique(model.labels_))
    bin['label_3'] = model.labels_
    plt.subplot(5, 10, i)
    plt.scatter(bin['WindSpeed'], bin['Power'], s=5, c=bin['label_3'], cmap='viridis')
    plt.xlabel('Wind Speed{}'.format(i))
    plt.ylabel('Power')
    # ax.set_xticks(np.arange(bin['WindSpeed'].min(), bin['WindSpeed'].max(), 0.5))
    plt.grid()
    data3 = pd.concat([data3, bin], axis=0, ignore_index=True)
plt.show()
```

```python
db_data = data_0.drop('label', axis=1).sort_values(by='WindSpeed').copy()
print(db_data['WindSpeed'].min(), db_data['WindSpeed'].max())
db_data['WindSpeed_bins'] = pd.cut(db_data['WindSpeed'], bins=list(np.arange(2.5, 18.5, 0.5)), labels=range(1, 32))

data3 = pd.DataFrame()
plt.figure(figsize=(30, 40))
for i in range(1, 32):
    # print('--------{}---------'.format(i))
    bin = db_data.loc[db_data['WindSpeed_bins'] == i, :]
    model = DBSCAN(eps=50, min_samples=5).fit(bin[['WindSpeed', 'Power']])
    # print('labels', i, np.unique(model.labels_))
    if i == 4:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 45 else 0)
    elif i in [5]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 100 else 0)
    elif i in [6]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 160 else 0)
    elif i in [7]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 210 else 0)
    elif i in [8]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 290 else 0)
    elif i in [9]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 350 else 0)
    elif i in [10]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 400 else 0)
    elif i in [11]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 500 else 0)
    elif i in [12]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 620 or x > 1300 else 0)
    elif i in [13]:
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 830 else 0)
    elif i in range(14, 26):
        bin['label_3'] = bin['Power'].apply(lambda x: 1 if x < 1100 else 0)
    else:
        bin['label_3'] = 0
    plt.subplot(5, 10, i)
    plt.scatter(bin['WindSpeed'], bin['Power'], s=5, c=bin['label_3'], cmap='viridis')
    plt.xlabel('Wind Speed')
    plt.ylabel('Power')
    plt.grid()
    data3 = pd.concat([data3, bin], axis=0, ignore_index=True)
plt.show()
```
